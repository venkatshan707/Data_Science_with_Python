The above was matplotlib style.

There is also an Object Oriented style of doing the same thing.

```python
# First we create a grid of plots
# ax will be an array of ywo Axes objects

fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
# First we create a grid of plots
# ax will be an array of ywo Axes objects

fig, ax = plt.subplots(2,2)
ax[0][0].plot(x, np.sin(x))
ax[1][1].plot(x, np.cos(x));
```

```python
# Adjusting the plot: line color and Styles

plt.plot(x, np.sin(x-0), color='blue') # specify color name
plt.plot(x, np.sin(x-1), color='g') # short color code (rgbcmyk) # https://cdn.firespring.com/images/1bc97d7d-1c72-4e08-9c54-f7e5d38d8d48.jpg
plt.plot(x, np.sin(x-2), color='0.75'); # Grayscale between 0 and 1, 0 will be dark grey, 1 will be light
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3)); # RGB tuple, values between 0 and 1
```

```python
# Adjust the line style

plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-') # solid
plt.plot(x, x+5, linestyle='--') # dashed
plt.plot(x, x+6, linestyle='-.') # dashdot
plt.plot(x, x+7, linestyle=':'); # dotted
```

```python
plt.plot(x, x+0, '-g') # solid green
plt.plot(x, x+1, '--c') # dashed cyan
plt.plot(x, x+2, '-.k') # dashdot black
plt.plot(x, x+3, ':r'); # dotted red
```
