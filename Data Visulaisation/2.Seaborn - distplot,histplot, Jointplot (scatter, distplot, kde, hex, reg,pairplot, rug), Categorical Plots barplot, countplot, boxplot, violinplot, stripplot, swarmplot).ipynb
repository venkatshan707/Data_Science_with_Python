__jointplot()__

It allows us to match up two distplots for bivariate data.

With our choice of a 'kind' parameter, we can give the following values:

>scatter

>reg

>resid

>kde

>hex

```python
sns.jointplot(x='total_bill', y='tip', data=tips); # by default, its scatter plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex');
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg'); # Regression Line
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde');
```

# Categorical Data Plots

__barplot() and countplot()__

barplot allows us to aggregate the categorical data off some function, by default the mean.

```python
sns.barplot(x='sex', y='total_bill', data=tips);
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.mean);
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std);
```

__countplot__

This is essentially the same as barplot except the estimator is explicitly counting the number of occurences, which is why we only pass the x value here.

```python
sns.countplot(x='sex', data=tips);
```

__boxplot__

It is used to identify the outliers in the data.

It is also known as Box and Whisker plot.

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow');
# x is a Categorical point
# y is a numeric point
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm');
# x is a Categorical point
# y is a numeric point
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='rainbow');
# x is a Categorical point
# y is a numeric point
```

People on Friday tend to have a larger bill if they are not Smoker verses if they are SMoker.

__violinplot__

It plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one or more categorical variables such that those distributions can be compared.

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow');
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow');
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set1');
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', split=True, data=tips, palette='Set1');
```

__stripplot__

```python
sns.stripplot(x='day', y='total_bill', data=tips);
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True);
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex');
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex',split= True);
```

__swarmplot = stripplot + violinplot__

```python
sns.swarmplot(x='day', y='total_bill', data=tips);
```

__Avoid using swarmplot for very large datasets because these points just go way to wise and get stacked on the top of each other.__

__Also most people might havent encountered it before.__

```python
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size=3);
# Use it only for EDA, and not for presentations to Clients or your manager
```

```python
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size=10);
```

```python
sns.violinplot(x='tip', y='day', data=tips, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=tips, color='black', size=1);
```

__clustermap__

It uses hierarchial clustering to produce a clustered version of the heatmap.

```python
sns.clustermap(pvflights);
```

```python
sns.clustermap(pvflights, cmap='coolwarm');
```

```python
sns.clustermap(pvflights, cmap='coolwarm', standard_scale=1);
```

Notice how the years and months are no longer in order, instead, they are grouped by similarity in value of passenger count. We can see that maximum people travel in July and Aug month, since they are both summer travel months.
