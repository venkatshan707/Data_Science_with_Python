# Combining Datasets: Merge and Join

```python
# One to One Join
df1 = pd.DataFrame({
    'employee' : ['Bob','Jake','Lisa','Sue'],
    'group' : ['Accounting','Engineering','Engineering','HR']
})
df2 = pd.DataFrame({
    'employee' : ['Lisa','Bob','Jake','Sue'],
    'hire_date' : [2004,2008,2012,2014]
})
print(df1)
print()
print(df2)
df3 = pd.merge(df1,df2)
df3
```

```python
# Many to One joins
df4 = pd.DataFrame({
    'group' : ['Accounting','Engineering','HR'],
    'supervisor' : ['Carly','Guido','Steve']
})
print(df3)
print()
print(df4)
pd.merge(df3,df4)
```

```python
# Many to Many Join

df5 = pd.DataFrame({
    'group' : ['Accounting','Accounting','Engineering','Engineering','HR','HR'],
    'skills' : ['math','spreadsheets','coding','linux','spreadsheets','organisation']
})
print(df1)
print()
print(df5)
pd.merge(df1,df5)
```

```python
print(df1)
print(df2)
pd.merge(df1,df2,on='employee')
```

```python
df3 = pd.DataFrame({
    'name' : ['Bob','Jake','Lisa','Sue'],
    'salary' : [70000,80000,120000,90000]
})
print(df1)
print()
print(df3)
x = pd.merge(df1,df3, left_on='employee', right_on='name')
x
```

```python
x.drop('name',axis=1,inplace=True)
```

```python
df3 = pd.DataFrame({
    'name' : ['Bob','Jake','Lisa','Sue'],
    'salary' : [70000,80000,120000,90000]
})
print(df1)
print()
print(df3)
pd.merge(df1,df3, left_on='employee', right_on='name').drop('name',axis=1)
```

__Join__

```python
#Inner and Outer join
df6 = pd.DataFrame({
    'name' : ['Peter','Paul','Mary'],
    'food' : ['fish','beans','bread']
})
df7 = pd.DataFrame({
    'name' : ['Mary','Joseph'],
    'drink' : ['wine','beer']
})
print(df6)
print()
print(df7)
pd.merge(df6,df7) # by default, this takes Inner join
```

```python
#Inner and Outer join
df6 = pd.DataFrame({
    'name' : ['Peter','Paul','Mary'],
    'food' : ['fish','beans','bread']
})
df7 = pd.DataFrame({
    'name' : ['Mary','Joseph'],
    'drink' : ['wine','beer']
})
print(df6)
print()
print(df7)
pd.merge(df6,df7, how='inner') # by default, this takes Inner join
```

```python
print(df6)
print()
print(df7)
pd.merge(df6,df7, how='outer')
```

```python
# How to apply a function over some column of the dataframe?
# If a student has <=35 marks and >=30 marks in Project Score, add a Grace of 5 marks
# to their score. Assuming that 35 is the passing marks.

def grace(x):
    if x>=30 and x<=35:
        x = x+5
        return x
    else:
        return x
```

```python
df['ProjectScore'] = df['ProjectScore'].apply(grace)
df
```

```python
# If the student has >=35 marks in ProjectScore, Recommend them with 'Yes'
df['Recommend'] = df['ProjectScore'].apply(lambda x: 'Yes' if x>=35 else 'No')
df
```

```python
# If the student has >=65 marks in ResearchScore, Recommend them with 'Yes'

def rec_fun(x):
    if x>=65:
        return 'Yes'
    else:
        return 'No'
```

```python
df['Recommend'] = df['ResearchScore'].apply(rec_fun)
df
```
